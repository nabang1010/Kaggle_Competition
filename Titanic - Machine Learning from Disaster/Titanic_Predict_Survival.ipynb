**Data Dictionary:**
* PassengerId: ID hành khách
* Pclass: Hạng vé 1-cao cấp, 2-hạng trung, 3-hạng thấp
* Name: Tên hành khách
* Sex: Giới tính
* Age: Tuổi
* SibSp: Xác định các mối quan hệ gia đình theo cách (Slibling - anh em ruột: anh em trai, chị em gái, anh chị em kế | Spouse - vợ chồng: chồng, vợ (tình nhân và hô phu đã bỏ))
* Parch: Xác định các mối quan hệ gia đình theo cách (Parent-phụ huynh: mẹ, cha | Child-con cái: con gái, con trai, con gái kế, con trai kế)
* Ticket: Số vé
* Fare: Giá vé
* Cabin: Số cabin
* Embarked: Cảng lên tàu
